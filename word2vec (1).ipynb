```python
from gensim.models import Word2Vec
import gensim
import nltk
from nltk import word_tokenize
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('/kaggle/working/viwik18/viwik18.txt', 'r') as f:
    data = f.read()
```

```python
data[:100]
```

```python
tokens = word_tokenize(data)
tokens[:20]
```

```python
tokens = [tokens]
tokens[0][:20]
```

```python
# Create CBOW model
model1 = gensim.models.Word2Vec(tokens, min_count=1,
                                vector_size=100, window=5)
```

```python
model1.save("word2vec.model")
```

```python
model1.wv['trang']
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

words_np = []
words_label = []
for word in model1.wv.index_to_key:
    words_np.append(model1.wv[word])
    words_label.append(word)

pca = PCA(n_components=2)
pca.fit(words_np)
reduced = pca.transform(words_np)
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,30)
for index,vec in enumerate(reduced):
    if index <100:
        x,y=vec[0],vec[1]
        plt.scatter(x,y)
        plt.annotate(words_label[index],xy=(x,y))
plt.show()
```
